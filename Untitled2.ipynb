```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import seaborn as sns
%matplotlib inline

print("Done")
```

```python
titanic=sns.load_dataset("titanic")
titanic
```

```python
sns.countplot(titanic['survived'])
plt.xlabel("survival")
plt.ylabel("count")
```

```python


# Draw a bar plot of survival by sex
sns.barplot(x="sex", y="survived", data=titanic, ci=None)
plt.title('Survival by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()

# Calculate percentages of females vs. males that survived
percentage_female_survived = titanic["survived"][titanic["sex"] == 'female'].mean() * 100
percentage_male_survived = titanic["survived"][titanic["sex"] == 'male'].mean() * 100

print("Percentage of females who survived:", percentage_female_survived)
print("Percentage of males who survived:", percentage_male_survived)
```

```python

sns.barplot(x="pclass", y="survived", data=titanic, ci=None)
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Calculate percentages of people by Pclass that survived
percentage_pclass_1_survived = titanic["survived"][titanic["pclass"] == 1].mean() * 100
percentage_pclass_2_survived = titanic["survived"][titanic["pclass"] == 2].mean() * 100
percentage_pclass_3_survived = titanic["survived"][titanic["pclass"] == 3].mean() * 100

print("Percentage of Pclass = 1 who survived:", percentage_pclass_1_survived)
print("Percentage of Pclass = 2 who survived:", percentage_pclass_2_survived)
print("Percentage of Pclass = 3 who survived:", percentage_pclass_3_survived)
```

```python


# Draw a bar plot of SibSp vs. survival
sns.barplot(x="sibsp", y="survived", data=titanic, ci=None)
plt.title('Survival by Number of Siblings/Spouses')
plt.xlabel('Number of Siblings/Spouses')
plt.ylabel('Survival Rate')
plt.show()

# Calculate percentages of individuals by SibSp that survived
for i in range(max(titanic["sibsp"]) + 1):
    percentage_sibsp_survived = titanic["survived"][titanic["sibsp"] == i].mean() * 100
    print(f"Percentage of SibSp = {i} who survived: {percentage_sibsp_survived:.2f}")
```

```python
#draw a bar plot for Parch vs. survival
sns.barplot(x="parch", y="survived", data=titanic)
plt.show()
```

```python
#sort the ages into logical categories
titanic["age"] = titanic["age"].fillna(-0.5)
bins = [-1, 0, 5, 12, 18, 24, 35, 60, np.inf]
labels = ['Unknown', 'Baby', 'Child', 'Teenager', 'Student', 'Young Adult', 'Adult', 'Senior']
titanic['ageGroup'] = pd.cut(titanic["age"], bins, labels = labels)

#draw a bar plot of Age vs. survival
sns.barplot(x="ageGroup", y="survived", data=titanic)
plt.show()
```
